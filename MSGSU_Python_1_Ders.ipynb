```python
#Dictionary Sözlük


#Sözlük yaratma yöntemleri:

Dict = {1:'Python', 2:'Egitim', 3:'Discord'}
print(Dict)

Dict2 = {'Ad': 'Hale', 1:[7,8,9]}
print(Dict2)

Dict3 = {1:'BMW', 2:'Audi', 3:{'A':'Selam', 'B': 'Naber', 'C': 'Ben'}}
print(Dict3)

Dict4= {}
print(Dict4)

Dict4[0] = 'Selam'
print(Dict4)

Dict4[2] = 'Naber'
print(Dict4)

Dict4['Sayilar'] = 2,3,5,7
print(Dict4)

Dict4['Egitim'] = {'1': 'Hale' , '2': 'Python'}
print(Dict4)

#Dictioanry içine erişim
print(Dict4.get(0)) #İçine key değerini yazarak erişiriz. Diğer değerlere erişmek için onların key değeri yazılır
print(Dict4['Sayilar']) #Get kullanmadan köşeli parantezler ile key değeri üzerinden erişim
print(Dict4['Egitim'])
```
